SAYISAL TÜREV

Bir f(x) fonksiyonun [a,b] tanım aralığında bir x noktasındaki türevi,



limit ifadesiyle tanımlanır. Eğer, f(x) in analitik ifadesi biliniyorsa, f' türevi içinde analitik bir ifade çoğu zaman bulunabilir. Ama bazen bu analitik türev alma işlemi çok karmaşık olabilir. Bazen de fonksiyon, deneysel ölçümlerde olduğu gibi, sadece belli noktalarda verilmiş olabilir. Bu durumlarda sayısal türev alma yoluna gidilir.

Sayısal türev problemi şöyle ortaya konur: f(x) fonksiyonu, eşit h aralıklarıyla sıralanmmış x0,x1,x2...xn noktalarında verilmiş olsun:


Buradaki küçük h değerine adım uzunluğu denir. Her xi noktasındaki f'(xi) türevi için sayısal hesaplamaya uygun yaklaşık bir ifade bulmak istiyoruz. Bunun için, fonksiyonun xi civarındaki Taylor açılımını yazalım:



Bu ifadeyi, aradığımız f'(xi) için çözersek,



Sağ taraftaki köşeli parantez içindeki terimler h değeriyle orantılı bir katkı verirler. O halde, h mertebesinde bir hatayla, sayısal türev ifadesi şöyle olur:



Bu ifadeye ileri fark türevi denir. O(h) terimi "h mertebesinden katkılar" adıyla bilinir. Benzer şekilde, geri fark türevi ifadesi de kolayca hesaplanabilir. Bunun için f(xi-h) nın Taylor açılımını yazarsak,



olur. Yine 1. türev terimini alır ve O(h) mertebesinden katkıları bir araya toplarsak,



bulunur. Hem ileri ve hem de geri fark türevlerinde hata payı h ile orantılıdır.

Daha iyi bir ifade bulabilir miyiz? Elbette, yukarıda verdiğimiz ileri ve geri Taylor açılımlarının farkını alalım:



Görüldüğü gibi, h^2 ile orantılı 2. türev terimleri birbirini götürmüştür. Buradan, aradığımız türevi yazarız:



o halde, h^2 mertebesinde bir hatayla, simetrik farklı türev ifadesi şöyle bulunur:



Bu ifadede h adım uzunluğu küçüldükçe, O(h^2) hata payı daha hızlı küçülür. Bu nedenle, simetrik türev ifadesi ileri farklı türevden daha iyi sonuç verir.

Şimdi, ikinci türev için bir ifade bulalım. Yukarıda, iki Taylor açılımı farkını almıştık. Bu kez, iki Taylor açılımı ifadelerinin toplamını alalım:



Birinci türev yokluğundan, ikinci türev ifadesini yalnız bırakırsak,



İkinci türev için, ileri ve geri farklı ifadeler, f(a±h) ve (a±2h) noktalarındaki Taylor açılımlarından elde edilir. Benzer şekilde işlem yapılırsa,



Bu ifadelerden görüldüğü gibi, aynı sayıda nokta kullanıldığı halde, simetrik farklı ifadenin hata payı daha küçük olmaktadır.

Bu yöntem, daha fazla nokta kullanarak, daha duyarlı türev ifadeleri geliştirilebilir. Bunun için, xi±2h noktalarındaki Taylor açılımları kullanılır ve gerekli toplama ve çıkarmalar yapılır. Örneğin, 5-nokta için simetrik farklı türev ifadeleri şöyle olur:



Fizik hesaplarında 3-noktalı ifadelerle hesaplamak yeterli olmaktadır. Daha yukarı duyarlılıktaki ifadeler gerektiğinde kullanılabilir.

Şimdi, tanjant fonksiyonunun x=1,2,3 noktalarındaki 1. türevini h=0.001 adım uzunluğu alarak hesaplayan bir program yazalım

```python
from math import *

def f(x):
    return tan(x)

def f1(x):
    return 1/cos(x)**2
h=0.001

for i in range(3):
    x=float(i+1)
    f1_ileri=(f(x+h)-f(x))/h
    f1_geri=(f(x)-f(x-h))/h
    f1_simetrik=(f(x+h)-f(x-h))/(2*h)
    f1_tam=f1(x)
    print(x,"%.10f"%f1_ileri, "%.10f"%f1_geri,"%.10f"%f1_simetrik,"%.10f"%f1_tam)
```

ADIM UZUNLUĞUNUN ETKİSİ

Sayısal türev alırken h adım uzunluğunu nasıl seçilmelidir? Hata payı O(h) veya O(h^2) ile orantılı olduğundan, büyük h değerleri seçilirse, hata büyük olacaktır. O halde yeterince küçük bir h değeri seçilmelidir.

Burada "yeterince küçük h" için kabaca şöyle bir kural vardır: Fonksiyonun ele alındığı aralık [a,b] ise, bu aralığın 1/100 veya 1/1000 kadar büyüklükte bir h değeri yeterli olur.

Fakat h değeri daha da küçük seçilirse, bu kez kesme hataları önem kazanır ve sonuçlar doğruluktan uzaklaşmaya başlar. Bunu görebilmek için, değişik h değerleri için aynı bir noktadaki türevin nasıl değiştiğine bakalım. Bu kez f(x)= eˣ fonksiyonunun x=1 noktasındaki simetrik 1. türevini h=0.1 değerinden başlayıp her defasında 10 kez azaltarak 12 kez hesaplayacak programı yazalım.

```python
from math import *

def f(x):
    return exp(x)
def f1(x):
    return exp(x)
h=0.1
x=1.0

for i in range(12):
    f1_simetrik=(f(x+h)-f(x-h))/(2*h)
    f1_tam=f1(x)
    fark=f1_tam-f1_simetrik
    print("%.12f"%h,"%.10f"%f1_simetrik,"%.10f"%f1_tam,"%.10f"%fark)
    h=h/10.0
```

Görüldüğü gibi, başlangıçta adım uzunluğu büyükken, sayısal türev ifadesinde O(h²) terimini atmaktan kaynaklanan kesme hatası önemli olmaktadır. Ancak, h küçüldükçe kesme hatası azalır.

Fakat, h değeri daha da küçüldüğünde, bu kez yuvarlama hatası önem kazanmaktadır, çünkü sayısal türev ifadesinde birbirine çok yakın iki sayının farkı alınmaktadır.
